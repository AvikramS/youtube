```python
import requests
import pandas as pd
```

```python
url= 'https://wmeng.exchanger.ru/asp/XMLWMList.asp?exchtype=34'
r = requests.get(url)
```

```python
import xml.etree.ElementTree as ET
import io
```

```python
xml_data = io.StringIO(r.text)
etree = ET.parse(xml_data)
```

```python
parsed= etree.getroot()
```

```python
amountin= []
amountout=[]
inoutrate=[] 
outinrate=[] 
procentbankrate=[]
allamountin=[]
querydate=[]
for info in parsed[1]:
    amountin.append(info.attrib.get('amountin'))
    amountout.append(info.attrib.get('amountout'))
    inoutrate.append(info.attrib.get('inoutrate'))
    outinrate.append(info.attrib.get('outinrate'))
    procentbankrate.append(info.attrib.get('procentbankrate'))
    allamountin.append(info.attrib.get('allamountin'))
    querydate.append(info.attrib.get('querydate'))
```

```python
dic= {'WMX': amountin, 'WMZ': amountout, 'WMX/WMZ': inoutrate, 'WMX/WMZ': outinrate, 'procentbankrate': procentbankrate, 'allamountin': allamountin, 'querydate':querydate}
BTCUSC= pd.DataFrame(dic)
BTCUSC.set_index('querydate', inplace= True)
```
